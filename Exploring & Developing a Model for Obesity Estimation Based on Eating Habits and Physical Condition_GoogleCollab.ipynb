### **Buiding the data**

```python
df_prep = df.copy()
# create dummy variables
df_prep = pd.get_dummies(df_prep,columns=["Gender","family_history_with_overweight",
                                          "FAVC","CAEC","SMOKE","SCC","CALC","MTRANS"])
df_prep.head()

# split dataset in features and target variable

# Features
X = df_prep.drop(columns=["NObeyesdad"])

# Target variable
y = df_prep['NObeyesdad']
# import sklearn packages for data treatments
from sklearn.model_selection import train_test_split # Import train_test_split function

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
X_train.columns
```

### **Feature Selection**

```python
from sklearn.feature_selection import SelectKBest, chi2
from numpy import array

select = SelectKBest(chi2, k=20)
X_train_new = select.fit_transform(X_train, y_train)
X_train_new.shape

filter = select.get_support()
features = array(X.columns)

print("All features:")
print(features)

print("Selected best 20:")
print(features[filter])
print(X_train_new)
```

### **Model Construction and Comparison**

### **Classification**

```python
df['NObeyesdad'].value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler


# Create instances of the models
random_forest = RandomForestClassifier(random_state=2020)
decision_tree = DecisionTreeClassifier()
knn = KNeighborsClassifier()
svm = SVC()
logistic_regression = LogisticRegression(max_iter=10000)

# Fit each model on the training set
random_forest.fit(X_train, y_train)
decision_tree.fit(X_train, y_train)
knn.fit(X_train, y_train)
svm.fit(X_train, y_train)
logistic_regression.fit(X_train, y_train)

# Predict the target variable for the testing set for each model
y_pred_rf = random_forest.predict(X_test)
y_pred_dt = decision_tree.predict(X_test)
y_pred_knn = knn.predict(X_test)
y_pred_svm = svm.predict(X_test)
y_pred_lr = logistic_regression.predict(X_test)

# Evaluate the performance of each model
accuracy_rf = random_forest.score(X_test, y_test)
accuracy_dt = decision_tree.score(X_test, y_test)
accuracy_knn = knn.score(X_test, y_test)
accuracy_svm = svm.score(X_test, y_test)
accuracy_lr = logistic_regression.score(X_test, y_test)

print("Random Forest Accuracy:", accuracy_rf)
print("Decision Tree Accuracy:", accuracy_dt)
print("KNN Accuracy:", accuracy_knn)
print("SVM Accuracy:", accuracy_svm)
print("Logistic Regression Accuracy:", accuracy_lr)
```
